```python
from pptx.util import Inches
from PIL import Image

def calculate_scaled_image_height(image_path, desired_width):
    ### Calculate the height of an image when scaled to a specific width.

    # Open image to get DPI
    with Image.open(image_path) as img:
        dpi = img.info.get('dpi', (96, 96))  # Default to 96 DPI if not found
        width_dpi, height_dpi = dpi
        img_width, img_height = img.size

    # Convert image dimensions from pixels to inches
    img_width_in_inches = img_width / width_dpi
    img_height_in_inches = img_height / height_dpi

    # Calculate the height based on the desired width
    scaled_height = (desired_width / img_width_in_inches) * img_height_in_inches

    return scaled_height

def calculate_vertical_position(image_path, desired_width, slide_height = 7.5):
    ### Calculate the vertical position to center an image within a shape.
        
    # Get the scaled height of the image
    scaled_height = calculate_scaled_image_height(image_path, desired_width)

    # Calculate vertical position for centering
    vertical_position = (slide_height - scaled_height) / 2

    return vertical_position
```
